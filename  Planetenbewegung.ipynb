```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

zeit = np.linspace(0, 100,10000)
# Wir betrachten eine Kraft in 1 / rÂ² mit unrealistischen Anfangsbedingungen
def dg_bewegung(mobil,zeit):
    x,y,dx,dy=mobil
    return [dx,dy,-x/(x**2+y**2)**1.5,-y/(x**2+y**2)**1.5]

X,Y,dX,dY=odeint(dg_bewegung, (1,0,0,0.5), zeit).T
plt.plot(X,Y)
plt.grid()
plt.show()
```
