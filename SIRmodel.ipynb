```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def deriv(y, t, beta, gamma):
    """
    y: Liste mit den 3 unbekannten Funktionen
    t: Zeit
    Beta, Gamma: die beiden Faktoren des Modells
    """
    S,I,R = y 

  # Beschreibung der 3 Differentialgleichungen
    dSdt = -S * I  * beta 
    dIdt = S * I  * beta  - gamma * I 
    dRdt = gamma * I 

    return dSdt, dIdt, dRdt 


# Zum Zeitpunkt t0, 70% gesund, 30% infiziert, 0 geheilt
y0 = 0.7, 0.3, 0

# Evolution über 28 Tage
t = np.linspace(0, 28)

# Modellparameter
beta = 0.5
gamma =  0.1

# Start der Integration von Differentialgleichungen
ret = odeint(deriv, y0, t, args = ( beta, gamma))
S,I,R = ret.T

plt.figure(figsize=(20,10))
plt.plot(t, S, label="Susceptible")
plt.plot(t, I,label="Infectious")
plt.plot(t, R, label="Recovered")

plt.xlabel("time")
plt.ylabel("number of individuals")
plt.legend()
plt.title(f"Proportion of individuals during an epidemic modeled with β = {beta} et γ = {gamma}")
```
