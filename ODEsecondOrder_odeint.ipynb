```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

a=-1
temps = np.linspace(0, 50,1000)
# Die Differentialgleichung als Funktion
def equation(Y,temps):
   # Wir zerlegen unser Y in (y, dy):
    (y,dy)=Y
  # Wir geben zur端ck, was y' ist:
    return [dy,a*y]

# Damit odeint die Werte von Y und Y 'getrennt zur端ckgibt, m端ssen Sie am Ende .T hinzuf端gen Transponierte Matrix
Y,dY=odeint(equation, [1,0], temps).T
plt.plot(temps,Y)
plt.show()
```
