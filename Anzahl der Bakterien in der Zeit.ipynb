```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint

## naive Algorithmusmethode
N_0 = 100  # Anfang Anzahl der Population
k = 0.2    # coef
times = np.arange(0,100,1) # Zeit
y  = []
N = N_0
# Iteration und Berechnung von N (t + 1)
for x in times:
  y.append(N)
  N  = N + N * k

plt.figure(figsize= (15,5))
plt.plot(times, y, "mo")
plt.xlabel("time")
plt.ylabel("number of bacteria")

## Algorthmische Methode mit Scipy

def deriv(N, t, K):
    "" "Derived from N with respect to time t and from the coefficient K" ""
    dNdt = K * N 
    return dNdt

N_0 = 100
k = 0.2
K = np.log(k + 1)
t = np.arange(0, 100, 1)
# Lösung der Differentialgleichung mit Odeint
ret = odeint(deriv, N, t, args = (K,) )

plt.figure(figsize= (15,5))
plt.xlabel("time")
plt.ylabel("Number of bacteria")
plt.plot(t, y, "go")
```
