```python
from scipy.integrate import solve_ivp
import numpy as np

# Systemkonstanten
R = 10  # Ohms
C = 0.01  # F
U = 5  # V

# Definition der Differentialgleichung
def equation(t, u):
    return (U-u)/(R*C)


t0 = 0
tf = 0.8  # sekunden
u0 = 0  # Anfangsbedingungen

# Résolution
voltage = solve_ivp(equation, [t0, tf], [u0])

#Berechnung des durch den Kondensator fließenden Stroms
current = (U-voltage.y[0])/R
```

```python
import matplotlib.pyplot as plt

# Plot der Spannung am Kondensator
plt.plot(voltage.t, voltage.y[0], label="u(t)")
plt.plot(voltage.t, current, label="i(t)")
plt.ylabel('u(t) en V et i(t) en A')
plt.xlabel("Time(s)")
plt.title("Voltage across the capacitor and current flowing through it")
plt.legend()
plt.show()
```
